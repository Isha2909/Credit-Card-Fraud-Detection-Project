```python
import optuna
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
from imblearn.over_sampling import SMOTE

# Define logistic regression model
logistic_regression = LogisticRegression()

def objective(trial):
    # Suggest hyperparameters
    C = trial.suggest_loguniform('C', 1e-5, 1e2)
    solver = trial.suggest_categorical('solver', ['liblinear', 'lbfgs'])
    if solver == 'lbfgs':
        penalty = 'l2'
    else:
        penalty = trial.suggest_categorical('penalty', ['l1', 'l2'])


    # Train logistic regression model
    model = LogisticRegression(C=C, solver=solver, penalty=penalty, random_state=42, max_iter=1000)
    model.fit(X_train, y_train)

    # Predict probabilities and calculate ROC AUC score
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    roc_auc = roc_auc_score(y_test, y_pred_prob)

    return roc_auc

# Create study and optimize
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

# Print best hyperparameters
print(f"Best hyperparameters: {study.best_params}")
print(f"Best ROC AUC: {study.best_value}")

# Train and evaluate the best model
best_params = study.best_params
best_model = LogisticRegression(**best_params, random_state=42, max_iter=1000)
best_model.fit(X_train, y_train)
y_pred_prob = best_model.predict_proba(X_test)[:, 1]
best_roc_auc = roc_auc_score(y_test, y_pred_prob)
print(f"Best model ROC AUC on validation set: {best_roc_auc}")
```

```python
from sklearn.tree import DecisionTreeClassifier
def objective(trial):
    # Suggest hyperparameters
    max_depth = trial.suggest_int('max_depth', 1, 32)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 20)
    max_features = trial.suggest_categorical('max_features', ['auto', 'sqrt', 'log2', None])

    # Train decision tree model
    model = DecisionTreeClassifier(
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        random_state=42
    )
    model.fit(X_train, y_train)

    # Predict probabilities and calculate ROC AUC score
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    roc_auc = roc_auc_score(y_test, y_pred_prob)

    return roc_auc

# Create study and optimize
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

# Print best hyperparameters
print(f"Best hyperparameters: {study.best_params}")
print(f"Best ROC AUC: {study.best_value}")

# Train and evaluate the best model
best_params = study.best_params
best_model = DecisionTreeClassifier(**best_params, random_state=42)
best_model.fit(X_train, y_train)
y_pred_prob = best_model.predict_proba(X_test)[:, 1]
best_roc_auc = roc_auc_score(y_test, y_pred_prob)
print(f"Best model ROC AUC on validation set: {best_roc_auc}")
```

```python
import optuna
from sklearn.ensemble import IsolationForest
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

# Define objective function for optimization
def objective(trial):
    # Define hyperparameters to optimize
    n_estimators = trial.suggest_int('n_estimators', 50, 200)
    max_samples = trial.suggest_categorical('max_samples', ['auto', 100, 200])
    max_features = trial.suggest_categorical('max_features', [1.0, 0.5, 0.7])

    # Create Isolation Forest model with current hyperparameters
    model = IsolationForest(
        n_estimators=n_estimators,
        max_samples=max_samples,
        max_features=max_features,
        random_state=42,
        contamination='auto'
    )

    # Train the model
    model.fit(X_train,y_train)

    # Predict on the validation set
    y_pred = model.predict(X_test)

    # Calculate ROC AUC score
    roc_auc = roc_auc_score(y_test, y_pred)

    return roc_auc

# Split the data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X_train,y_train, test_size=0.2, random_state=42)

# Create study object and optimize hyperparameters
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

# Get the best hyperparameters
best_params = study.best_params
print("Best Hyperparameters:", best_params)

```
